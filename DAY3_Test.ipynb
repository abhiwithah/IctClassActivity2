```python
import pandas as pd
import numpy as np
df = pd.read_csv('auto-mpg.csv')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
#NUMPY problems
#Basic Array Operations
mpg_arr = df['mpg'].to_numpy()
mean_mpg = np.mean(mpg_arr)
median_mpg = np.median(mpg_arr)
std_mpg = np.std(mpg_arr)
count_high_mpg = np.sum(mpg_arr > 25)

print(f"Mean: {mean_mpg:.2f}, Median: {median_mpg}, Std Dev: {std_mpg:.2f}")
print(f"Number of cars with mpg > 25: {count_high_mpg}")
```

```python
#Filtering
cyl_arr = df['cylinders'].to_numpy()
name_arr = df['car name'].to_numpy()
cars_above_6_cyl = name_arr[cyl_arr > 6].tolist()

print(f"Cars with > 6 cylinders: {cars_above_6_cyl}")
```

```python
#Broadcasting
hp_arr = df['horsepower'].to_numpy()
hp_mean = np.nanmean(hp_arr)
hp_arr[np.isnan(hp_arr)] = hp_mean
updated_hp = hp_arr * 1.10

print("Updated Horsepower (first 5):", updated_hp[:5])
```

```python
#Boolean Indexing
displacement_arr = df['displacement'].to_numpy()
origin_arr = df['origin'].to_numpy()
avg_disp_europe = np.mean(displacement_arr[origin_arr == 2])

print(f"Average Displacement for origin 2: {avg_disp_europe:.2f}")
```

```python
#Conditional Aggregates
unique_cyls = np.unique(cyl_arr)
cyl_mpg_means = {cyl: np.mean(mpg_arr[cyl_arr == cyl]) for cyl in unique_cyls}

print("Mean MPG by cylinders:")
for cyl, mean_val in cyl_mpg_means.items():
    print(f"{cyl} cylinders: {mean_val:.2f}")
```

```python
#PANDAS problem
#Basic Exploration
print("First 10 rows:")
print(df.head(10))
print(f"\nRows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\nSummary Statistics:")
print(df.describe())
```

```python
#Filtering and Indexing
filtered_cars = df[(df['model year'] == 75) & (df['weight'] < 3000)]
print(filtered_cars[['car name', 'weight', 'mpg']])
```

```python
#Handling Missing Data
print("Missing values per column:\n", df.isnull().sum())
hp_median = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(hp_median)
```

```python
#Data Transformation
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']

print(df[['car name', 'power_to_weight_ratio']].head())
```

```python
#Group By
origin_mpg_mean = df.groupby('origin')['mpg'].mean()

print(origin_mpg_mean)
```

```python
#Visualization Preparation
summary_df = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()

print(summary_df)
```

```python
#Exporting Data
high_mpg_cars = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]
high_mpg_cars.to_csv('high_mpg_cars.csv', index=False)
print("File 'high_mpg_cars.csv' has been created successfully.")
```
